```python
from pythainlp import word_vector
model = word_vector.WordVector(model_name="thai2fit_wv").get_model()
```

```python
import pandas as pd

thai2dict = {word: model[word] for word in model.index_to_key}
thai2vec = pd.DataFrame.from_dict(thai2dict, orient='index')
thai2vec.head()
```

```python
from sklearn.manifold import TSNE
import pickle
tsne = TSNE(n_components=2, init='pca', perplexity=30, learning_rate='auto')
thai2plot = tsne.fit_transform(thai2vec)
with open("thai2plot.pkl", "wb") as f:
    pickle.dump(thai2plot, f)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# สร้าง labels จาก model
labels = model.index_to_key

# ฟังก์ชัน plot
def plot_with_labels(low_dim_embs, labels, filename=None, figsize=(10,10), axis_lims=None):
    import matplotlib.pyplot as plt
    import matplotlib.font_manager as fm
    plt.figure(figsize=figsize)
    font = fm.FontProperties(fname="TH-Sarabun-New-Regular.ttf", size=14)
    for i, label in enumerate(labels):
        x, y = low_dim_embs[i]
        plt.scatter(x, y)
        plt.annotate(label, fontproperties=font, xy=(x, y), xytext=(5, 2),
                     textcoords='offset points', ha='right', va='bottom')
    if axis_lims: plt.axis(axis_lims)
    if filename: plt.savefig(filename)
    plt.show()

# แสดงคำในช่วง index 100–200
plot_with_labels(thai2plot[200:500], labels[200:500])
```

```python
import re

thai_letters = 'กขฃคฅฆงจฉชซฌญฎฏฐฑฒณดตถทธนบปผฝพฟภมยรฤฤๅลฦฦๅวศษสหฬอฮะัาำิีึืุูเแโใไ็่้๊๋์'
WORDS = {w: i for i, w in enumerate(model.index_to_key)}


def correction(word):
    def known(words): return set(w for w in words if w in WORDS)
    def edits1(word):
        splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
        deletes = [L + R[1:] for L, R in splits if R]
        transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]
        replaces = [L + c + R[1:] for L, R in splits if R for c in thai_letters]
        inserts = [L + c + R for L, R in splits for c in thai_letters]
        return set(deletes + transposes + replaces + inserts)
    def edits2(word): return (e2 for e1 in edits1(word) for e2 in edits1(e1))
    def candidates(word): return known([word]) or known(edits1(word)) or known(edits2(word)) or [word]
    def P(w): return -WORDS.get(w, 0)
    return max(candidates(word), key=P)

print(correction("พัดนา"))  # → พัฒนา
print(correction("ทำไมถึงมาโรงเรียนสาน"))   # → จ้า
```

```python
from pythainlp.tokenize import word_tokenize
import numpy as np

def get_novel_vector(text, model):
    tokens = word_tokenize(text, keep_whitespace=False)
    vectors = [model[word] for word in tokens if word in model]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)
```

```python
# สมมติเรามีนิยาย 3 เรื่อง
novels = [
    {"title": "รักต้องห้ามขององค์ชาย", "desc": "ความรักระหว่างเจ้าชายกับหญิงสามัญชน"},
    {"title": "โลกคู่ขนาน", "desc": "การเดินทางข้ามเวลาเพื่อเปลี่ยนแปลงอดีต"},
    {"title": "ซาตานกับนักบวช", "desc": "เรื่องราวแห่งศรัทธาและการล่อลวง"}
]

# แปลงเป็นเวกเตอร์
for novel in novels:
    novel["vector"] = get_novel_vector(novel["title"] + " " + novel["desc"], model)

# คำนวณความคล้าย
from sklearn.metrics.pairwise import cosine_similarity

target = novels[0]["vector"]
similarities = [cosine_similarity([target], [n["vector"]])[0][0] for n in novels]

# แนะนำ
ranked = sorted(zip(novels, similarities), key=lambda x: x[1], reverse=True)
for novel, score in ranked[1:]:  # ข้ามเรื่องต้นฉบับ
    print(novel["title"], f"({score:.2f})")
```

```python
def get_vector_from_tags(tags: list[str], model):
    vectors = [model[tag] for tag in tags if tag in model]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)
```

```python
# tag ที่ผู้ใช้กดอ่านบ่อย
user_tags = ['โรแมนติก', 'ย้อนยุค', 'แฟนตาซี']

user_vector = get_vector_from_tags(user_tags, model)

# เทียบกับนิยายทั้งหมด
for novel in novels:
    novel["score"] = cosine_similarity([user_vector], [novel["vector"]])[0][0]

novels.sort(key=lambda n: n["score"], reverse=True)
for n in novels:
    print(n["title"], f"({n['score']:.2f})")
```

```python
def recommend_by_keyword(keyword: str, all_novels: list, model, topk=5):
    query_vec = get_novel_vector(keyword, model)
    results = []
    for novel in all_novels:
        sim = cosine_similarity([query_vec], [novel["vector"]])[0][0]
        results.append((novel["title"], sim))
    return sorted(results, key=lambda x: x[1], reverse=True)[:topk]

# ทดลองค้นหา
recommend_by_keyword("เจ้าหญิงต้องคำสาป", novels, model)
```
